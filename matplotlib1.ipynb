```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import numpy as np
```

```python
x=np.arange(1,11)
y=np.arange(11,21)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c='r') #scattering
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
plt.savefig('mtplt.png')
```

```python
y=x*x*x
```

```python
y
```

```python
plt.plot(x,y,'ro',linestyle='dashed',linewidth=1,markersize=8)  #line plot(third arguement null or simple r g b)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(y,x,'r*')
```

```python
#bar plots

x1=[2,3,4,24]
y1=[6,7,8,9]

x2=[3,5,25,12]
y2=[4,13,12,16]
plt.bar(x1,y1,color='g')
plt.bar(x2,y2,color='r')
```

```python

#histogram

a=np.array([22,33,45,2,23,22,5,5,2,5,2,55,22,24,11,5])
plt.hist(a,bins=20)
```

```python
#boxplot using matplotlib

data = [np.random.normal(0,std,100) for std in range(1,4)]
data
plt.boxplot(data)
```
